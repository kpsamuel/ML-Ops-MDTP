### Linear Regression Model implementation

```python
x_features = df.drop("Chance of Admit", axis = 1)
y_predict = df['Chance of Admit']
```

```python
from sklearn.model_selection import train_test_split
from sklearn import preprocessing
from sklearn.preprocessing import StandardScaler,MinMaxScaler
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import pickle
```

```python
scaler = MinMaxScaler()
x_train, x_test, y_train, y_test = train_test_split(x_features, y_predict, test_size = 0.2, random_state = 42)

x_train = scaler.fit_transform(x_train)
x_train = pd.DataFrame(data=x_train, columns=x_features.columns)

x_test = scaler.transform(x_test)
x_test = pd.DataFrame(data=x_test, columns=x_features.columns)

print("training data : {} | test data : {}".format((x_train.shape, y_train.shape), (x_test.shape, y_test.shape)))
```

```python
model = LinearRegression()
model.fit(x_train, y_train)
predictions = model.predict(x_test)

rms=np.sqrt(mean_squared_error(y_test, predictions))
train_score = model.score(x_train,y_train)
test_score = model.score(x_test, y_test)

print("RMS : {} | train score : {} | test score : {}".format(rms, train_score, test_score))
```

```python
with open("sampledataset/linear/linearRegressionModel.pkl", "wb") as mp:
    pickle.dump(model, mp)
    
with open("sampledataset/linear/trainedFeatureScaler.pkl", "wb") as mp:
    pickle.dump(scaler, mp)
    
```

### model explainability

```python
import shap
shap.initjs()
```

#### Explain a single prediction from the test set

```python
x_train_summary = shap.kmeans(x_train,k=len(x_train.columns))
exp = shap.KernelExplainer(model=model.predict, data=x_train_summary)
```

```python
data_point = 0
shap_values = exp.shap_values(x_test.iloc[data_point])
shap_values
```

```python
shap.force_plot(exp.expected_value, shap_values, x_test.iloc[data_point])
```

```python
model.predict(np.array(x_test.iloc[data_point]).reshape(1, -1))
```

### Explain all the predictions in the test set

```python
x_test_shap_values = exp.shap_values(x_test)
shap.summary_plot(x_test_shap_values, x_test)
```
